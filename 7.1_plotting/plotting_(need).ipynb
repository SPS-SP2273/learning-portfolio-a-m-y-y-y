```python
plt.style.use("default")

# plt.xkcd()
time=np.linspace(0,10)
y2 = time**2
sin_t = np.sin(time)

plt.plot(time, y2, label="$y^2$", c="orange")
plt.ylabel("$y^2$", rotation=0)
plt.legend()

another_y = plt.twinx() # create a new y-axis
plt.plot(time, sin_t, label=r"$\sin x$")
plt.ylabel(r"$\sin x$", rotation=0)

# plt.scatter(time, y)
plt.xlabel("time (s)")
plt.title("My First Plot")
plt.legend()
plt.grid(alpha=.25)
plt.savefig("hahha.png", dpi=300)
```
