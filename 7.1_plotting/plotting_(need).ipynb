```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
time = np.arange(10)
y2=time**2
y3=time**3

# use colorbrewer2.org to find suitable colours
plt.plot(time, y2, label="$y^2$", linestyle = "--", color = "#1b9e77")
plt.plot(time, y3, label="$y^3$", color = "#7570b3")
#plt.scatter(time,y)
plt.xlabel("time (s)")
plt.ylabel("Hahaha (ha)")
plt.title("My First Plot")
plt.legend(loc='upper left')
plt.grid(alpha=.25)
```

## 2.1 Adding mathematical functions

## We need another axis!

```python
plt.style.use("default")

# plt.xkcd()
time=np.linspace(0,10)
y2 = time**2
sin_t = np.sin(time)

plt.plot(time, y2, label="$y^2$", c="orange")
plt.ylabel("$y^2$", rotation=0)
plt.legend()

another_y = plt.twinx() # create a new y-axis
plt.plot(time, sin_t, label=r"$\sin x$")
plt.ylabel(r"$\sin x$", rotation=0)

# plt.scatter(time, y)
plt.xlabel("time (s)")
plt.title("My First Plot")
plt.legend()
plt.grid(alpha=.25)
plt.savefig("hahha.png", dpi=300)
```
